# importing libraries and data

```python
import numpy as np
import pandas as pd
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
file_path = r'C:\Users\kunal\OneDrive\Desktop\Projects\coded\nextword-predictor\next_word_predictor.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    text_data = file.read()

text_data = text_data.lower().replace('\n', ' ')
print(text_data[:500])
```

# tokenization and embedding

```python
vocab_size = 10000
tok = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tok.fit_on_texts([text_data])
seqs = tok.texts_to_sequences([text_data])[0]
```

```python
x = []
y = []
for i in range(1, len(seqs)):
    x.append(seqs[:i])
    y.append(seqs[i])
```

```python
padded_x = pad_sequences(x, maxlen=20, padding='pre')
```

```python
y = np.array(y)
```

```python
embed_dim = 20
rnn_units = 32
```

# building the model

```python
model = keras.Sequential()
model.add(
    layers.Embedding(input_dim=vocab_size,output_dim=embed_dim)
    )
model.add(layers.LSTM(32))
model.add(layers.Dense(vocab_size, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(
    padded_x, y, 
    epochs=10,
    batch_size = 30,
    validation_split=0.1
)
```

# saving the model


```python
model.save('model.keras')
```
